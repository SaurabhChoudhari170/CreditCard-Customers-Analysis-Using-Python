#### Creating AGE GROUP Column OF Customers

```python
labels=["{0} - {1}".format(i, i + 19) for i in range(1, 80, 20)]   # list compression 
df['age_group'] = pd.cut(df.CustoMaleer_Age
, range(1, 100, 20), right=False, labels=labels)
df["age_group"].value_counts()
```
